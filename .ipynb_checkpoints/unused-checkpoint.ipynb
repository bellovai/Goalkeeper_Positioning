```python
def create_dataframe(directory):
    
    Return dataframe of all json files in directory
    
    
    # Create empty lists
    ls_shot = []
    ls_possession_team = []
    ls_location = []

    # Iterate over all files in directory
    for filename in os.listdir(directory):

        # Save filepath to current file
        filepath = os.path.join(directory, filename)
        # If current file has json ending
        if filename.split('.')[1] == 'json':
            # Transform current json file to dataframe
            df_file = pd.read_json(filepath)
            # Transform columns of current dataframe to lists and add them to existing lists
            ls_shot.extend(df_file['shot'].tolist())
            ls_possession_team.extend(df_file['possession_team'].tolist())
            ls_location.extend(df_file['location'].tolist())

    # Create empty dataframe and add complete lists 
    df = pd.DataFrame()
    df['shot'] = ls_shot
    df['possession_team'] = ls_possession_team
    df['location'] = ls_location

    # Return dataframe
    return df
```

```python
def implement_knn_model(df):
    
    Return dataframe with TOGKP from knn model
    
    
    # Create previous and men dataframe
    df = add_alternative_data(df)
    # Create empty lists
    ls_x_gk_knn = []
    ls_y_gk_knn = []
    
    # Iterate over all rows in dataframe
    for row_i in range(len(df)):
        
        # Define shot position
        x_shot = df.at[row_i, 'x_shot']
        y_shot = df.at[row_i, 'y_shot']
        # Create empty list
        ls_distance = []

        if 'Women' not in df.at[row_i, 'possession_team']:
            
            # Iterate over all rows in dataframe again
            for row_j in range(len(df)):

                if 'Women' not in df.at[row_j, 'possession_team']:
                    # Define similar shot position
                    x_shot_similar = df.at[row_j, 'x_shot']
                    y_shot_similar = df.at[row_j, 'y_shot']
                    # Add distance between shot position and similar shot position to list
                    ls_distance.append(math.dist([x_shot, y_shot], [x_shot_similar, y_shot_similar]))
                    
        else:
            
            # Iterate over all rows in dataframe again
            for row_j in range(len(df)):
                
                if 'Women' in df.at[row_j, 'possession_team']:
                    # Define similar shot position
                    x_shot_similar = df.at[row_j, 'x_shot']
                    y_shot_similar = df.at[row_j, 'y_shot']
                    # Add distance between shot position and similar shot position to list
                    ls_distance.append(math.dist([x_shot, y_shot], [x_shot_similar, y_shot_similar]))
                  
        # Sort indices by ascending distance values and save value 1 to 21 (i.e. don't save the smallest)
        sorted_indices = np.argsort(ls_distance).tolist()[1:21]

        # Define counter variables
        x_gk_tot = 0
        y_gk_tot = 0
        # Sum up gk coordinates from 20 most similar shots
        for similar_shot in sorted_indices:
            x_gk_tot = x_gk_tot + df.at[similar_shot, 'x_gk']
            y_gk_tot = y_gk_tot + df.at[similar_shot, 'y_gk']
        # Add mean gk coordinates from 20 most similar shots to list
        ls_x_gk_knn.append(round(x_gk_tot/20, 2))
        ls_y_gk_knn.append(round(y_gk_tot/20, 2))    
        
    # Add lists to dataframe
    df['x_gk_knn'] = ls_x_gk_knn
    df['y_gk_knn'] = ls_y_gk_knn
    
    # Return dataframe
    return df
```

```python
# Define method
def myplot(x, y, s, bins=1000):
    # Method
    heatmap, xedges, yedges = np.histogram2d(x, y, bins=bins)
    heatmap = gaussian_filter(heatmap, sigma=s)
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    # Return
    return heatmap.T, extent

fig, axs = plt.subplots(2, 2)
sigmas = [0, 16, 32, 64]

for ax, s in zip(axs.flatten(), sigmas):
    if s == 0:
        ax.plot(x, y, 'k.', markersize=5)
        ax.set_title('Scatter plot')
    else:
        img, extent = myplot(x, y, s)
        ax.imshow(img, extent=extent, origin='lower', cmap=cm.jet)
        ax.set_title('Smoothing with  $\sigma$ = %d' % s)
        
# Show plot
plt.show()
```

### Evaluate Statsbomb xG Performance

```python
# Copy dataframe
df = df_1v1.copy()
# Create empty lists
ls_outcome_bin = []
ls_xg_bin = []

# Encode outcome to binary
for row in range(len(df)):
    if df.at[row, 'outcome'] == 'Goal':
        ls_outcome_bin.append(int(1))
    else:
        ls_outcome_bin.append(int(0))

# Encode xg to binary        
for row in range(len(df)):
    if df.at[row, 'statsbomb_xg'] >= 0.5:
        ls_xg_bin.append(int(1))
    else:
        ls_xg_bin.append(int(0))

# Add lists do dataframe
df['outcome_bin'] = ls_outcome_bin
df['statsbomb_xg_bin'] = ls_xg_bin

# Save dataframe
df = df[['outcome', 'outcome_bin', 'statsbomb_xg', 'statsbomb_xg_bin']]

# Create variables
tn = 0
tp = 0
fn = 0
fp = 0

# Update variables
for row in range(len(df)):
    if df.at[row, 'outcome_bin'] == 0 and df.at[row, 'statsbomb_xg_bin'] == 0:
        tn = tn + 1
    elif df.at[row, 'outcome_bin'] == 1 and df.at[row, 'statsbomb_xg_bin'] == 1:
        tp = tp + 1
    elif df.at[row, 'outcome_bin'] == 1 and df.at[row, 'statsbomb_xg_bin'] == 0:
        fn = fn + 1
    elif df.at[row, 'outcome_bin'] == 0 and df.at[row, 'statsbomb_xg_bin'] == 1:
        fp = fp + 1

# Calculate metrics
accuracy = round((tp + tn)/(tp + tn + fn + fp), 2)
precision = round(tp/(tp + fp), 2)
recall = round(tp/(tp + fn), 2)
f1 = round(2*precision*recall/(precision + recall), 2)

# Print metrics
print('Accuracy:', accuracy)
print()
print('Precision:', precision)
print()
print('Recall:', recall)
print()
print('F1:', f1)
```

```python
def one_hot_encode(df, col):

    # Define encoder
    encoder = OneHotEncoder(dtype = 'int', handle_unknown = 'ignore')
    # Encode column
    df_encoder = pd.DataFrame(encoder.fit_transform(df[[col]]).toarray())
    
    # Create list with categorical values
    ls_cat_vals = sorted(df[col].unique().tolist())
    # Rename columns
    df_encoder.columns = ls_cat_vals
    
    # Merge encoded columns with original dataframe
    df = df.join(df_encoder)
    # Delete shot location column
    del df[col]

    # Return dataframe
    return df
```
